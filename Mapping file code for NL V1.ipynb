```python
import pandas as pd
from zipfile import ZipFile
import os
import xlwings as xw
def xlsb_read(filex):
    app = xw.App()
    book = xw.Book(filex)
    sheet = book.sheets('OUT')
    df = sheet.range('A1').options(pd.DataFrame, expand='table').value
    df.reset_index(inplace = True)
    book.close()
    app.kill()
    return df
main_path = input('Path of zip files :')
extract_path = input('Path of extraction : ')
os.chdir(main_path)
for file in os.listdir():
    abc = ZipFile(file)
    abc.extractall(path = extract_path)
    print(file, ' extracted')
os.chdir(extract_path)
for file2 in os.listdir():
    if 'xlsb' in file2:
        if 'FCA' in file2:
            df = xlsb_read(file2)
            df.fillna('NA',inplace = True)
            df.columns = ["Input_Item","LOCAL MANUFACTURER","Local Brand","Need Occasion 1","Functional claim 1","Product Form 1","Functional Claim 2","Flavour Fragrance 1","Packaging Usage","Functional Claim 3","Need Occasion 2","Flavour Fragrance 2","Pack size"]
            df.to_excel('Mapping_FCA.xlsx',index=False)
            print(file2, ' Created')
        elif 'MKO' in file2:
            df1 = xlsb_read(file2)
            df1.fillna('NA',inplace = True)
            df1.columns = ["Input_Item","LOCAL MANUFACTURER","Local Brand","Need Occasion 2"]
            df1.to_excel('Mapping_MKO.xlsx',index=False)
            print(file2, ' Created')
    else:
        file_read = pd.read_excel(file2)
        file_read.fillna('NA',inplace = True)
        if 'HCA' in file2:
            file_read.columns = ["Input_Item","LOCAL MANUFACTURER","Local Brand","Product Form 1","Functional Claim 3","Packaging Usage","Flavour Fragrance 1","Functional claim 1","Functional Claim 2","Flavour Fragrance 2","Pack size","Need Occasion 2","Need Occasion 1"]
            file_read.to_excel('Mapping '+file2,index=False)
            print(file2, 'Created')
        elif 'HCO' in file2:
            file_read.columns = ["Input_Item","LOCAL MANUFACTURER","Local Brand","Product Form 1","Packaging Usage","Flavour Fragrance 1","Functional claim 1","Need Occasion 1","Pack size","Flavour Fragrance 2","Need Occasion 2","Functional Claim 3","Functional Claim 2"]
            file_read.to_excel('Mapping '+file2,index=False)
            print(file2, 'Created')
        elif 'MAE' in file2:
            file_read.columns = ["Input_Item","LOCAL MANUFACTURER","Local Brand","Need Occasion 1","Pack size","Product Form 1","Flavour Fragrance 1","Functional claim 1","Need Occasion 2","Functional Claim 2","Functional Claim 3","Packaging Usage","Flavour Fragrance 2"]
            file_read.to_excel('Mapping '+file2,index=False)
            print(file2, 'Created')
        elif 'MKF' in file2:
            file_read.columns = ["Input_Item","LOCAL MANUFACTURER","Local Brand","Product Form 1","Flavour Fragrance 1","Need Occasion 1","Functional Claim 2","Functional claim 1","Functional Claim 3","Flavour Fragrance 2","Pack size","Packaging Usage"]
            file_read.to_excel('Mapping '+file2,index=False)
            print(file2, 'Created')
        elif 'MKL' in file2:
            file_read.columns = ["Input_Item","LOCAL MANUFACTURER","Local Brand","Pack size","Product Form 1","Packaging Usage","Flavour Fragrance 1","Need Occasion 2","Functional claim 1","Functional Claim 2","Functional Claim 3","Flavour Fragrance 2"]
            file_read.to_excel('Mapping '+file2,index=False)
            print(file2, 'Created')
        elif 'MKN' in file2:
            file_read.columns = ["Input_Item","LOCAL MANUFACTURER","Local Brand","Need Occasion 2","Pack size","Need Occasion 1","Product Form 1","Flavour Fragrance 1","Functional claim 1","Packaging Usage","Functional Claim 3","Functional Claim 2","Flavour Fragrance 2"]
            file_read.to_excel('Mapping '+file2,index=False)
            print(file2, 'Created')
        else:
            print('Wrong file collection')
for file3 in os.listdir():
    if 'Mapping' not in file3:
        os.remove(file3)
print('Mapping File creation is completed')
```
